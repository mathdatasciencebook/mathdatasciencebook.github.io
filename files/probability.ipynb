```

from matplotlib.pyplot import *
from numpy.random import default_rng as rng
samples = rng().binomial

p, N = .5, 100000

for n in [5,15,50,500]:
	data = samples(n,p,N)
	hist(data, bins = n, edgecolor = 'k')
	grid()
	show()
```

```

from sklearn import datasets

iris = datasets.load_iris()

dataset = iris['data']
iris['feature_names']
```

```

petal_lengths = dataset[:,2]
mean(petal_lengths)
```

```

hist(petal_lengths, bins = 50)

grid()
show()
```

```

# n = minibatch size

def minibatch_mean(n):
	rng().shuffle(petal_lengths)
	return mean(petal_lengths[:n])
	
minibatch_mean(5)
```

```

n, N = 1, 100000
	
Xbar = [ minibatch_mean(n) for _ in range(N) ]
hist(Xbar, bins = 50)

grid()
show()
```

```

figure(figsize=(8,4)) 
# three subplots
rows, cols = 1, 3
	
N = 100000
	
for i,n in enumerate([3,5,20], start = 1):
	Xbar = [ minibatch_mean(n) for _ in range(N) ]
	subplot(rows,cols,i)
	grid()
	hist(Xbar, bins = 50)
	
show()
```

```

def sums(dataset, k):
	if k == 1: return dataset
	else: 
		s = sums(dataset, k-1)
		return array([ a + s for a in dataset ]).reshape(-1)

dataset = array([ -11.2,  1.4, 5 ])

for k in range(1, 20):
	s = sums(dataset, k)
	hist(s, bins = 50, edgecolor = 'k')
	title(f'sum size {k}')
	grid()
	show()
```
