```

from numpy import *
set_printoptions(legacy = '1.25')

# vectors
u = array([1,2,3,4,5])
v = array([6,7,8,9,10])
w = array([11,12,13,14,15])

# arrange as columns
A = column_stack([u,v,w])

r, s, t = 1, 2, 3
x = array([r,s,t])

dot(A,x) == r*u + s*v + t*w
```

```

from scipy.linalg import orth
	
# returns minimal orthonormal spanning set 
# for column space of A, as columns
	
orth(A)
```

```

# returns minimal orthonormal spanning set 
# for row space of A, as columns

orth(A.T)
```

```

from scipy.linalg import null_space

# returns minimal orthonormal spanning set 
# for nullspace of A, as columns
null_space(A)
```

```

# returns minimal orthonormal spanning set 
# for nullspace of A.T, as columns	
null_space(A.T)
```

```

from sympy import *
	
# convert to sympy matrix
Asym  = Matrix(A)
	
Q = Asym * Asym.T
evects = Q.eigenvects()
u1 = 260 * evects[2][2][0]
u2 = 260 * evects[1][2][0]
null =	   evects[0][2]
U = Matrix.hstack(u1, u2, *null)
	
Q = Asym.T * Asym
evects = Q.eigenvects()
v1 = 94 * evects[2][2][0]
v2 = 94 * evects[1][2][0]
null =	  evects[0][2]
V = Matrix.hstack(v1, v2, *null)
	
display(U, V)
```
