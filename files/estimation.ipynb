```

from numpy import *
set_printoptions(legacy = '1.25')
from numpy.random import default_rng as rng

def Angle(u,v):
	a = dot(u,v)
	b = dot(u,u)
	c = dot(v,v)
	theta = arccos(a / sqrt(b*c))
	return degrees(theta)

samples = rng().normal

mu, sdev, d = 0, 1, 784

for _ in range(20):
	u = samples(mu,sdev,d)
	v = samples(mu,sdev,d)
	print(Angle(u,v))
```

```

samples = rng().binomial

n, p, d = 1, .5, 784 # one coin toss
# n, p, d = 3, .5, 784 # three coin tosses

for _ in range(20):
	u = samples(n,p,d)
	v = samples(n,p,d)
	print(Angle(u,v))
```
