```

from pandas import *
from numpy import *

mnist = read_csv('mnist.csv').to_numpy()

# separate rows into data and labels
# first column is the labels
labels = mnist[:,0]
# all other columns are the pixels
dataset = mnist[:,1:]

shape(mnist), shape(dataset), shape(labels)
```

```

tuple = (28,28)
pixels = dataset[1].reshape(tuple)
```

```

from matplotlib.pyplot import * 

scatter(2, 3, s = 50)

grid()
show()
```

```

for i in range(28):
	for j in range(28): 
		scatter(i, j, s = pixels[i,j], edgecolor = 'k')

grid()
show()
```

```

from matplotlib.pyplot import *

imshow(pixels, cmap = 'gray_r') 

grid()
show()
```

```

import matplotlib.pyplot as plt
	
plt.imshow(pixels, cmap = 'gray_r') 

plt.grid()
plt.show()
```

```

from matplotlib.pyplot import imshow, grid, show
	
imshow(pixels, cmap = 'gray_r') 

grid()
show()
```

```

i = j = range(28)
i, j = meshgrid(i, j)

scatter(i, j, s = pixels, c = sin(i*j), edgecolor = 'k', cmap = 'hsv')

grid()
show()
```
