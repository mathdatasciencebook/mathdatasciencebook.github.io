```

from numpy import *
set_printoptions(legacy = '1.25')

from numpy.random import default_rng as rng
samples = rng().binomial
	
n, p, N = 5, .5, 10
	
# counting heads from n tosses sampled N times
samples(n,p,N)
```

```

from scipy.stats import binom

k, n, p = 5, 10, .5

B = binom(n,p)

# probability of k heads
B.pmf(k)
```

```

from scipy.special import comb

# code to verify binomial pmf

def f(n,k,p): 
	return binom(n,p).pmf(k)
def g(n,k,p): 
	return comb(n,k,exact = True) * p**k * (1-p)**(n-k)

n, p = 10, .5

pmf1 = array([ f(n,k,p) for k in range(n+1) ]) 
pmf2 = array([ g(n,k,p) for k in range(n+1) ])
	
allclose(pmf1,pmf2)
```

```

from scipy.stats import entropy
from matplotlib.pyplot import *

def H(p): return entropy([p,1-p])
def stirling(n, p): 
	return exp(n * H(p)) / sqrt(2*n*pi * p*(1-p))

# can't divide by p = 0 in stirling, so start from .01
p = arange(.01, 1, .01)

n = 10
plot(p, comb(n, n*p), label = 'binomial coefficient')
plot(p, stirling(n, p), label = 'entropy approximation')

tex = f'$n = {n}$'
title(f'number of tosses {tex}', usetex = True)
legend()

grid()
show()
```

```

n, k = 10, 7

def f(p): return (n+1) * binom(n,p).pmf(k)
	
p = arange(0, 1, .01)
plot(p, f(p), c = 'blue', lw = .5)

grid()
show()
```
