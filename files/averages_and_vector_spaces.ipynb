```

from numpy.random import default_rng as rng
samples = rng().random
	
N, d = 20, 2
	
# Nxd array
dataset = samples((N,d))
mu = mean(dataset, axis = 0)
	
scatter(*mu, c = 'red')
scatter(*dataset.T, c = 'blue')
	
grid()
show()
```
