```

from pandas import *
from numpy import *
from numpy.linalg import matrix_rank
set_printoptions(legacy = '1.25')

mnist = read_csv('mnist.csv').to_numpy()
dataset = mnist[:,1:]

mu = mean(dataset, axis = 0)
A = dataset - mu

matrix_rank(A)
```

```

set_printoptions(legacy = '1.25')

def find_first_defect(A):
	N, d = shape(A)
	previous = 0
	for n in 1 + arange(N):
		r = matrix_rank(A[:n])
		print((r,n), end = ', ')
		if r == previous or r == d: break
		previous = r
		
find_first_defect(A)
```

```

from matplotlib.pyplot import *

# A is centered dataset matrix
rmax = matrix_rank(A)

def dimension_staircase(A,rmax):
	N, d = shape(A)
	staircase = [ ]
	for n in 1 + arange(N):
		r = matrix_rank(A[:n])
		print(r, end = ', ')
		staircase.append(r)
		if r == rmax: break
	return staircase, n

# for MNIST, this may take several hours or days
staircase, n = dimension_staircase(A, rmax)

title(f'number of vectors = {n}, rank = {rmax}')
stairs(staircase, range(n+1), linewidth = 1, color = 'r')
grid()
show()
```
