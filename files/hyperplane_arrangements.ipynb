```python
from numpy import *
from matplotlib.pyplot import *

weights = array([[1,0,1],[0,1,1],[1,1,-2]])
d = len(weights) 

cmap = colormaps['hsv']
colors = cmap(linspace(0,1,d))

from matplotlib.colors import rgb2hex

for c in colors: print(rgb2hex(c))
```

```python
def level(w,x): return w[0]*x[0] + w[1]*x[1] + w[2]
```

```python
def maxcolor(x, weights, colors):
    y = array([ level(w,x) for w in weights ])
    i =argmax(y)
    return rgb2hex(colors[i])
    

N = 10
s = t = linspace(0,1,N)

x = array(meshgrid(s, t))
x = moveaxis(x,0,2)
shape(x)

```

```python

cc = array([ [ maxcolor(x[i,j,:], weights, colors) for i in range(N) ] for j in range(N) ]) 
shape(cc)
```
