```

def row(i,d):
	v = [0]*d
	v[i] = 2
	if i > 0: v[i-1] = -1
	if i < d-1: v[i+1] = -1
	if i == 0: v[d-1] += -1
	if i == d-1: v[0] += -1
	return v

from numpy import *

def Q(d): return array([ row(i,d) for i in range(d) ])
```

```

from scipy.linalg import eigh
from matplotlib.pyplot import *

d = 50
E = eigh(Q(d))[0]
stairs(E, range(d+1), label = 'numpy')

k = arange(d)
lamda = 2 - 2*cos(2*pi*k/d)
sorted = sort(lamda)

scatter(k, lamda, s = 5, label = 'unordered')
scatter(k, sorted, c = 'red', s = 5, label = 'increasing order')

grid()
legend()
show()
```

```

lamda = arange(0.01, 3.99, .01)
density = 1/(pi*sqrt(lamda*(4-lamda)))
plot(lamda, density, label = 'eigen density', c = 'b', lw = .5, ls = '-')
plot(lamda, (lamda/2)*density, label = 'trace density', c = 'r', lw = .5, ls = '--')

# r'...' means raw string
frac = r'$\displaystyle\frac'
denom = r'{\pi\sqrt{\lambda(4-\lambda)}}$'
eigen = frac + '{1}' + denom
trace = frac + r'{\lambda/2}' + denom
text(.5,0.6,eigen,usetex = True, size = 'large', c = 'b')
text(.5,.45,trace,usetex = True, size = 'large', c = 'r')

ylim(top = 0.7)
grid()
legend()
show()
```
