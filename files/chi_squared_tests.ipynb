```
 
from numpy import *
set_printoptions(legacy = '1.25')
from scipy.stats import chi2 as U

# observed and expected are arrays
# array arithmetic is element-wise
	
def goodness_of_fit(observed, expected):
	d, n = len(observed), sum(observed)
	u = -n + sum(observed**2 / expected)
	return 1 - U(d-1).cdf(u)

```

```
 
observed = array([17,12,14,20,29,28])
expected = array([20,20,20,20,20,20])

p = goodness_of_fit(observed, expected)
print(f'pvalue: {p}')
```

```

# observed and expected are arrays
# array arithmetic is element-wise

def chi2_independence(observed):
	n = sum(observed)
	N, d = shape(observed)
	rowsum = sum(observed, axis= 1)
	colsum = sum(observed, axis= 0)
	expected = outer(rowsum, colsum) / n
	u = -n + sum(observed**2 / expected)
	deg = (N-1)*(d-1)
	return 1 - U(deg).cdf(u)
	
observed = array([[68,56,32],[52,72,20]])
p = chi2_independence(observed)
print(f'pvalue: {p}')
```
