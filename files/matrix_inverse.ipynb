```

from numpy import *
set_printoptions(legacy = '1.25')

A = array([[1,2],[3,4]])
x = array([5,6])
b = dot(A,x)

shape(A) == (len(b), len(x))
```

```

from scipy.linalg import inv

# solving Ax = b
x = dot(inv(A), b)
x
```

```


# vectors
u = array([1,2,3,4,5])
v = array([6,7,8,9,10])
w = array([11,12,13,14,15])
	
# arrange as columns
A = column_stack([u,v,w])
```

```

from scipy.linalg import pinv

Aplus = pinv(A)

round(Aplus, 2), round(150*Aplus, 2)
```
