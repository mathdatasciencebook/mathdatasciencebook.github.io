```

from scipy.integrate import quad

d = 2

def f(x): return x**d

a, b = -1, 1

# quad returns integral plus error of f(x) 
quad(f,a,b)
```

```

def plot_and_integrate(f, a, b, pi_ticks = False):
	# draw x-axis and y-axis
	axhline(0, c = 'k', lw = 1)
	axvline(0, c = 'k', lw = 1)
	# set x-axis ticks at multiples of pi/2
	if pi_ticks: set_pi_ticks(a,b)
	x = linspace(a, b, 100)
	plot(x, f(x))
	positive = f(x) >= 0
	negative = f(x) < 0
	fill_between(x, f(x), 0, color = 'g', where = positive, alpha = .5)
	fill_between(x, f(x), 0, color = 'r', where = negative, alpha = .5)
	# just the integral, no error estimate
	I = quad(f, a, b, limit = 1000)[0]
	title(f'integral equals {I}', fontsize = 10)
	grid()
	show()
```

```

def f(x): return x**2
a, b = -1, 1
plot_and_integrate(f, a, b)

def f(x): return sin(x)
a, b = 0, 3*pi
plot_and_integrate(f, a, b, pi_ticks = True)

def f(x): return sin(x)/x
a, b = 0.001, 3*pi
plot_and_integrate(f, a, b, pi_ticks = True)
```
