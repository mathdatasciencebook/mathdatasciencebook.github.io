```python
from numpy import *
from matplotlib.pyplot import *
from numpy.random import default_rng as rng


M = array([ [1,2], [4,3], [5,6],[7,8]],dtype=float)
B = array([1,2,3,4],dtype=float)
# center M and B
d = 4
#M = rng().random((d,2))
M -= mean(M,axis = 0)
#B = rng().random(d)
B -= mean(B)

colors = ['red','green','blue','yellow']

data0 = {} # points x-coord
data1 = {} # points y-coord

for color in colors:
    data0[color] = [ ]
    data1[color] = [ ]

a, A = -10,10
delta = 0.05

for x in arange(a,A, delta):
    for y in arange(a,A, delta):
        X = array([x,y])
        Y = array ([ dot(m,X) + b for m,b in zip(M,B) ] )
        ymax = max(Y)
        for i in range(d):
            if ymax == Y[i]: 
                data0[colors[i]].append(x)
                data1[colors[i]].append(y)

xlim(a,A)
ylim(a,A)

for i,color in enumerate(colors):
    scatter(data0[color],data1[color],c=color)
    m, b = M[i], B[i]
    if m[1]: plot([a,A],[-(b-m[0]*a)/m[1],-(b+m[0]*A)/m[1]],lw=0.5,c='white')
    else: plot([-(b-m[1]*a)/m[0],-(b+m[1]*A)/m[0]],[a,A],lw=0.5,c='white')

print(M,B)

show()
    
    


```

```python


colors = ['red','green']

phat = qhat = rhat = .333
n = 10

def U(n,p,q):
    r = 1 - p - q
    if p < 0.01: p = 0.01
    if q < 0.01: q = 0.01
    if r < 0.01: r = 0.01
    return n * ((p-phat)**2/p + (q-qhat)**2/q + (r-rhat)**2/r)

N = 20

p = q = linspace(0,1,N)
p, q = meshgrid(p, q)

for i in range(N):
    for j in range(N):
        if p[i,j] + q[i,j] < 1:
            if U(n,p[i,j],q[i,j]) < 5: scatter(p[i,j],q[i,j],c='red',s=1)
            else: scatter(p[i,j],q[i,j],c='green',s=1)

show()
    
    

```
