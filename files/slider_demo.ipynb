```python
%matplotlib ipympl

from numpy import *
from matplotlib.pyplot import *
from matplotlib.widgets import Slider

close()

ax = axes()

# Initial plot data
a,b = -10, 10
x = linspace(a,b,100)
amp = 0

def f(a,x): return x**2 + a*sin(x)
    
line, = ax.plot(x, f(a,x))

# Create the slider
axfreq = axes([0.25, 0.1, 0.65, 0.03])
slider = Slider(axfreq, 'Amplitude', 0, 30.0)

# Define the update function
def update(val):
    a = slider.val
    line.set_ydata(f(a,x))
    ax.canvas.draw_idle() # Redraw the plot

# Connect the slider to the update function
slider.on_changed(update)

ax.grid()
show()
```
