```

from sympy import *

a, b, c = symbols('a, b, c')
eq1 = a + 2*b + c - 1
eq2 = a - b - (a-c)*(a+b)
solutions = solve([eq1,eq2], a, b)

solutions
 
```

```

from scipy.stats import binom
	
n, p = 8, .5
B = binom(n,p)
	
for k in range(n+1): print(k, B.pmf(k), B.cdf(k))
```

```

from scipy.stats import poisson
	
lamda = 1
P = poisson(lamda)
	
for k in range(10): print(k, P.pmf(k), P.cdf(k))
```
