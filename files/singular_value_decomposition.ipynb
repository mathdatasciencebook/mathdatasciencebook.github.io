```

from numpy import *
from scipy.linalg import svd
set_printoptions(legacy = '1.25')

u = array([1,2,3,4,5])
v = array([6,7,8,9,10])
w = array([11,12,13,14,15])

A = column_stack([u,v,w])
U, sigma, Vt = svd(A)
V = Vt.T # svd returns V.T not V

U, sigma, V
```

```

# build diag matrix S
	
p = min(shape(A))
S = zeros(shape(A))
S[:p,:p] = diag(sigma)
	
allclose(A, dot(U, dot(S, V.T)))
```
