```

from numpy import *
set_printoptions(legacy = '1.25')

from scipy.linalg import pinv

# projection of  vector b
# onto column space of A

# assume len(b) == len(A)

def project(A,b):
	Aplus = pinv(A)
	x = dot(Aplus,b)    # reduced
	return dot(A,x)     # projected
```

```

# projection of column vector b
# onto column space of U
# with orthonormal columns
	
# assume len(b) == len(U)

def project_to_ortho(U,b):
	x = dot(U.T,b)      # reduced
	return dot(U,x)     # projected
```

```

def gs(A):
	N, d = shape(A)
	U = zeros((N,0))
	# step through columns
	for b in A.T:
		v = project_to_ortho(U, b)
		if not allclose(b, v):
			u = (b-v) / sqrt(dot(b-v, b-v))
			U = column_stack([U, u])
	return U
```
