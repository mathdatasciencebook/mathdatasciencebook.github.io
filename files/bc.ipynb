```python
from numpy.random import default_rng as rng
samples = rng().random

r, s, t = (3,1,2), (), (5,3,7,1)

u = broadcast_shapes(r,s,t)
	
a, b, c = samples(r), samples(s), samples(t)
arrays = [a,b,c]

d, e, f = broadcast_arrays(a, b, c)
	
u == (5,3,7,2) and  shape(d) == u and shape(e) == u and shape(f) == u
```

```python
def bs(*shapes):
    d = max([ len(s) for s in shapes ])
    n = tuple( max([ s[-i-1] for s in shapes if len(s) > i ]) for i in range(d) )
    for s in shapes:
        for i in range(d):
            if i < len(s) and s[-i-1] != 1 and s[-i-1] != n[i]: return 'not broadcastable'
    return n[::-1]

shapes = (3,1,2), (7,2), (5,3,7,1)

bs(*shapes)
```

```python



def ba(*arrays):
    n = bs(*[shape(a) for a in arrays])
    if n == 'not broadcastable': return n
    d = len(n)
    new = []
    for a in arrays:
        for i in range(d): 
            s = shape(a)
            if i >= len(s): a = array([ a for _ in range(n[-i-1]) ] )
            elif s[-i-1] == n[-i-1]: pass
            elif s[-i-1] == 1:
                a = moveaxis(a,-i-1,0)
                a = array([ a[0] for _ in range(n[-i-1]) ])
                a = moveaxis(a,0,-i-1)
            else: return 'not broadcastable' 
        new.append(a)
    return new
    
d1, e1, f1 = ba(a,b,c)

all(d == d1) and all(e == e1) and all(f == f1)  
#shape(d1), shape(e1), shape(f1), 
```
