```python

from pandas import *
from numpy import *
from numpy.linalg import matrix_rank
set_printoptions(legacy = '1.25')

mnist = read_csv('mnist.csv').to_numpy()
dataset = mnist[:,1:]
```

```python
from mpmath import *

dps = 200
mpf_array = frompyfunc(mpf,1,1)
dataset = mpf_array(dataset)

mu = mean(dataset, axis = 0)
A = dataset - mu
```

```python
from sklearn.preprocessing import StandardScaler

# standardize dataset
A = StandardScaler().fit_transform(A)
```

```python
matrix_rank(A)
```

```python
# iterating over rows!!
# assume first row b nonzero

def allclose(x, y, rtol=mpf(1.e-5), atol=mpf(1.e-8)):
    return all(less_equal(abs(x-y), atol + rtol * abs(y)))

atol = mpf(1.e-8)

def dimension_staircase(A):
    A, b = A[1:], A[0]
    N, d = shape(A)
    u = b / sqrt(dot(b,b))
    U = u.reshape((1,d))
    r, staircase = 0, [ ]
    for n in range(N):
        A, b = A[1:], A[0]
        v  = dot(U.T, dot(U, b))
        distance = sqrt(dot(b-v,b-v))
        if distance > atol:
        #if not allclose(b,v):
            r += 1
            u = (b-v) / distance
            U = vstack([U,u])
        print(r, end = ', ')
        staircase.append(r)
    return staircase, n+1
```

```python
# mpmath

staircase, n = dimension_staircase(A[:600])

title(f'number of vectors = {n+1}')
stairs(staircase, range(n+1), linewidth = 1, color = 'r')
grid()
show()
```
