```

from scipy.special import factorial

n = 10
factorial(n, exact = True)
```

```

def perm_tuples(n,k):
	if k == 0: return [ () ]
	else: return [ (i,*t) for i in range(n) for t in perm_tuples(n,k-1) if i not in t ]
	
perm_tuples(5,2)
```

```

from numpy import *
set_printoptions(legacy = '1.25')

from scipy.special import perm

n, k = 5, 2	
perm(n, k,exact = True)
```

```

perm(n,k, exact = True) == len(perm_tuples(n,k))
```

```

def comb_tuples(n,k):
	if k == 0: return [ () ]
	elif k == 1: return perm_tuples(n,1)
	else: return [  (i, *t) for i in range(n) for t in comb_tuples(n,k-1) if i < t[0] ] 
	
comb_tuples(5,2)  
```

```

[ (0,1), (0,2), (0,3), (0,4), (1,2), (1,3), (1,4), (2,3), (2,4), (3,4) ]
```

```

from scipy.special import comb
	
n, k = 5, 2
comb(n, k, exact = True)
```

```

comb(n,k, exact = True) == len(comb_tuples(n,k))
```
