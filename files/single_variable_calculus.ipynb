```

from sympy import diff, symbols
from scipy.special import factorial

def sym_legendre(n):
	# symbolic variable
	x = symbols('x')
	# symbolic function
	p = (x**2 - 1)**n
	nfact = factorial(n, exact = True)
	# symbolic nth derivative
	return p.diff(x,n)/(nfact * 2**n)
```

```

from sympy import simplify

for n in range(6): 
	display(simplify(sym_legendre(n)))
```

```

from sympy import lambdify
	
def num_legendre(n):
	x = symbols('x')
	f = sym_legendre(n)
	return lambdify(x, f, 'numpy')	
```

```

# r'...' is a raw string. In a raw string
# the backslash \ is not an escape character

def label(k): 
	if k == 0: return '$0$'
	elif k == 1: return r'$\pi/2$'
	elif k == -1: return r'$-\pi/2$'
	elif k == 2: return r'$\pi$'
	elif k == -2: return r'$-\pi$'
	elif k%2 == 0: return f'${k//2}' +  r'\pi$'
	else: return f'${k}' +  r'\pi/2$'
	
	
def set_pi_ticks(a,b):
	base = pi/2
	m = floor(b/base)
	n = ceil(a/base)
	k = arange(n,m+1, dtype = int)
	veclabel = vectorize(label)
	return xticks(k*base, veclabel(k))
```
